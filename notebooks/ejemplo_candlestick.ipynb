```python
import yfinance as yf
from src.TimeSeries import TimeSeries
from src.TimeSeriesPlot import TimeSeriesPlot
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-muted')
```

```python
meli = yf.download('MELI', start='2024-08-31')
dates = meli.index
meli_prices = meli['Close']['MELI']
meli_high = meli['High']['MELI']
meli_low = meli['Low']['MELI']
meli_open = meli['Open']['MELI']
```

```python
ts = TimeSeries(dates, meli_prices, meli_open, meli_high, meli_low)

plot = TimeSeriesPlot(ts)
plot.add_candlestick()
plot.add_grid()
plot.show()
```

```python
plot = TimeSeriesPlot(ts)
plot.add_candlestick()
plot.add_ema(20)
plot.add_ema(100)
plot.add_grid()
plot.add_grid()
plot.show()
```
